## Labelling a plot

```python
pyplot.plot([2,3,8,4])
pyplot.xlabel("indices of y values")
pyplot.ylabel('actual data which is plotted')
pyplot.show()
```

## Drawing graph lines

```python
pyplot.plot([1,2,3,4],[1,4,9,16])
pyplot.grid()
pyplot.xlabel('nums')
pyplot.ylabel('squares')
pyplot.show()
```

## Dotted plot

```python
pyplot.plot([1,2,3,4],[1,4,9,16], 'ro') # ro = red dot
pyplot.grid()
pyplot.xlabel('nums')
pyplot.ylabel('squares')
pyplot.show()
```

## plotting multiple plots on single figure

```python
t = numpy.arange(0,5,0.2)
pyplot.plot(t,t**2,'b--',label='^2') # blue ---
pyplot.plot(t,t**2.2, 'rs--',label = '^2.2') # red square with ---
pyplot.plot(t,t**2.5,'g^',label = '^2.5') # green triangle
pyplot.plot(t,t**2.8,'yo',label='^2.8') # yellow dot
pyplot.legend(loc = 'upper right') # shows hint to read graph.
# loc determines location on which hint is shown
pyplot.grid()
pyplot.show()
```

```python
t = numpy.arange(0,5,0.5)
color_shape = 'r--'
line_width = 5
pyplot.plot(t,t**2,color_shape, linewidth = line_width)
pyplot.show()
```

## Lines object
plot() function returns list of all .Line2D object which are plotted

using setproperty method we can configure line

its keyword args is used to set style of line


```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
# plot() function returns lines object

lines = pyplot.plot(x1,y1,x2,y2)
pyplot.setp(lines[0], color='b', linewidth = 2.0)
pyplot.setp(lines[1], color='g', linewidth = 4)
pyplot.show()
```

## Plotting on Subplots

```python
def func(t):
    return(numpy.exp(-t) * numpy.cos(2 * numpy.pi*t))

t1 = numpy.arange(0, 5, 0.1)
t2 = numpy.arange(0, 5, 0.3)

pyplot.figure(1)

pyplot.subplot(211) # rows|columns|working
pyplot.grid()
pyplot.plot(t1,func(t1),'b--',label='t1')
pyplot.legend()

pyplot.subplot(212)
pyplot.grid()
pyplot.plot(t2,func(t2),'go',label ='t2')
pyplot.legend()
pyplot.show()
```

## plots used in Data analytics 
- scatter plot
- bar plot
- histogram
- boxplot
- pair plot
- heat map

to deal with these plots, we use matplotlib.pyplot and seaborn
**Scatter plot** - used for bivariate analysis

**Histogram** - used for univariate analysis. shows frequency of continuous data

**BarPlot** - same as Histogram, but for discrete data

### Scatter plot


```python
cars_data = pandas.read_csv('files\\Toyota.csv',index_col=0, na_values=['??','????'])

# dropna is used to drop null rows or columns as per axis
cars_data.dropna(axis=0,inplace = True)
cars_data.isna().sum()

```

```python
pyplot.scatter(cars_data['Age'], cars_data['Price'],c = 'green')
pyplot.title('scatter plot of price vs age')
pyplot.xlabel('Age')
pyplot.ylabel('Price')
pyplot.show()
```

For overlapping values, scatterplot is not a good representation of univariate data. so we move on to Histogram.
### Histogram


```python
pyplot.hist(cars_data['KM'],color='skyblue',edgecolor='black',bins = 5)
# bins = no. of columns
# returns array of frequency and eight.
```

### Bar plot

```python
langs = ['C','C++','Java','Python']
students = [22,34,32,94]
pyplot.bar(langs,students)
pyplot.show()

# plotting categorical data from dataframe
pyplot.bar(cars_data['FuelType'].value_counts().index[0:],cars_data['FuelType'].value_counts(),color='green')

print(cars_data['FuelType'].value_counts())
pyplot.show()
```
