# Seaborn

```python
import seaborn
import pandas
import matplotlib.pyplot as plt
import numpy
```

```python
cars_data = pandas.read_csv('files\\Toyota.csv', index_col = 0, na_values = ['??', '????'])
```

```python
cars_data.dropna(axis = 0, inplace = True)
cars_data.isna().sum()
```

## Generating a scatter plot in seaborn

```python
seaborn.set(style = 'darkgrid')
# price vs age
# by default it gives a regression line
seaborn.regplot(x = cars_data['Age'], y = cars_data['Price'])
```

```python
seaborn.regplot(x = cars_data['Age'], y = cars_data['Price'],fit_reg = False,marker = '*')
```

```python
# age vs price by fuel type
seaborn.lmplot(x = 'Age', y = 'Price', data = cars_data, fit_reg = False, hue = 'FuelType', legend = True, palette = 'Set1')
```

```python
# histogram in seaborn
# histogram is used for continuous data
x = cars_data['Age']
seaborn.distplot(x,kde=False,bins = 5,color = 'r')
```

```python
# bar plot is used for discrete data
seaborn.countplot(x = 'FuelType', data=cars_data,hue = 'Automatic')
```
